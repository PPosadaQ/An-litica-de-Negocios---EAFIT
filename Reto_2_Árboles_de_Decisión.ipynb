**Árboles de Decisión**
0. Cargar las librerías de referencia para el modelo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #Graficador
from sklearn.tree import DecisionTreeClassifier
```

1. Se procede con la carga de los datos del archivo Travel_Plan_Score

```python
nxl= '/content/2. BD1_Travel_Data_int.xlsx'
XDB = pd.read_excel(nxl, sheet_name = 0)
XDB = XDB.dropna() #Elimina celdas vacías o sin datos
XDB.head(100)
```

```python
#Se procede con la selección de las variables de trabajo
XD=np.array(XDB[['College','FoodSpend','Income','TravelSpend']]) #Datos de Entrada
print(XD)
yd=np.array(XDB[['CreditCard']]) #Este es el dato que quiero que el modelo pronostique
print(yd)
```

2. Creación del modelo - Decision Tree

```python
mar=DecisionTreeClassifier(criterion='gini', max_depth = 4) #4 Niveles o variables de entrada
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Evaluamos el modelo
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]

Ex= (VN+VP)/(VN+VP+FN+FP) #Que porcentaje de los datos pronosticó correctamente (Es VP y VN)
print("La exactitud alcanzada por el modelo es:",Ex)

Se= VP/(FN+VP) #Como le fue con los positivos (Es FN y VP) Total Aprobados
print("La sensibilidad alcanzada por el modelo fue:",Se)

Es= VN/(VN+FP) #Como le fue con los negativos (Es VN y FP) Total Prenegados
print("La especificidad alcanzada por el modelo es:",Es)
```

3. ¿Y Donde está el Árbol?

```python
#Estas son las librerías para graficar el árbol
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

VS = ['College','FoodSpend','Income','TravelSpend'] #Son los títulos para los nodos
dot_graph=export_graphviz(mar,feature_names=VS)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbol.png')
```

4. Se procede con el pronóstico de los nuevos solicitantes

```python
XDB2=pd.read_excel(nxl, sheet_name=1)
XDB2 = XDB2.dropna() #Elimina las celdas vacías o sin datos
XDB2.head(100)
```

```python
XD2= np.array(XDB2[['College','FoodSpend','Income','TravelSpend']]) #Datos de entrada
print(XD2)

#Pronosticamos la PreAprobación
ydp2=mar.predict(XD2)
print(ydp2)
print("El número de PreAprobados es:" ,np.sum(ydp2))
print("El numero total de Datos es:" , len(ydp2))
print("El numero de PreNegados es:" , len(ydp2)-np.sum(ydp2))
```
